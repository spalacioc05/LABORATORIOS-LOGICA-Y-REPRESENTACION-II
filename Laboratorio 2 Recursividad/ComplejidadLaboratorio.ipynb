```python
# 1. Determine el funcionamiento del algoritmo.
# 2. Calcule la complejidad computacional.

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def detect_duplicates(arr):
    sorted_arr = insertion_sort(arr)

    duplicates = []
    for i in range(1, len(sorted_arr)):
        if sorted_arr[i] == sorted_arr[i - 1]:
            duplicates.append(sorted_arr[i])

    return duplicates

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
duplicates = detect_duplicates(numbers)
print("Números duplicados:", duplicates)
```

```python
# 1. Determine el funcionamiento del algoritmo.
# 2. Calcule la complejidad computacional.

def remove_target_sum_subarrays(arr, target):
    i = 0
    while i < len(arr):
        j = i
        current_sum = 0
        while j < len(arr):
            current_sum += arr[j]
            if current_sum == target:
                del arr[i:j + 1]
                i = max(0, i - 1)
                break
            j += 1
        else:
            i += 1
    return arr

numbers = [1, 2, 3, 7, 5, 2, 1, 2, 2, 6]
target_sum = 5
result = remove_target_sum_subarrays(numbers, target_sum)
print("Lista después de eliminar subarrays:", result)
```

```python
# Problema: Detección de Picos en Señales de Sensores
# El objetivo es encontrar los picos (puntos máximos locales) en un conjunto de lecturas de un sensor.
# Un pico es un punto que es mayor que sus vecinos.

# 1. Calcule la complejidad computacional.
# 2. Optimice el algoritmo.



def detect_peaks_naive(data):
    peaks = []
    n = len(data)
    for i in range(1, n - 1):

        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            peaks.append((i, data[i]))
    return peaks

sensor_data = [1, 3, 7, 1, 2, 6, 3, 4, 5, 0, 8]
peaks = detect_peaks_naive(sensor_data)
print("Picos detectados:", peaks)
```
