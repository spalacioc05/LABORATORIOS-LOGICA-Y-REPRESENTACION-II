```python
# 1. Determine el funcionamiento del algoritmo.
# 2. Calcule la complejidad computacional.

def insertion_sort(arr):
    for i in range(1, len(arr)): #O(n)
        key = arr[i] #O(n)
        j = i - 1 #O(n)
        while j >= 0 and key < arr[j]: #O(n * (n + 1))
            arr[j + 1] = arr[j] #O(n * n)
            j -= 1 #O(n * n)
        arr[j + 1] = key #O(n)
    return arr #O(1)

def detect_duplicates(arr):
    sorted_arr = insertion_sort(arr) #O(n^2)

    duplicates = [] #O(1)
    for i in range(1, len(sorted_arr)): #O(n)
        if sorted_arr[i] == sorted_arr[i - 1]: #O(n)
            duplicates.append(sorted_arr[i]) #O(n)

    return duplicates #O(1)

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3] #O(1)
duplicates = detect_duplicates(numbers) #O(n^2)
print("Números duplicados:", duplicates) #O(1)
```

1. Funcionamiento del Algoritmo

    insertion_sort:

    Entrada: Una lista arr.
    Proceso: Ordena la lista utilizando el algoritmo de ordenación por inserción.
    Recorre la lista desde el segundo elemento hasta el final.
    Para cada elemento, lo compara con los elementos anteriores y lo inserta en la posición correcta.
    Salida: La lista ordenada.

    detect_duplicates:

    Entrada: Una lista arr.
    Proceso:
    Ordena la lista utilizando insertion_sort.
    Recorre la lista ordenada y detecta elementos duplicados.
    Si encuentra elementos duplicados, los añade a una lista duplicates.
    Salida: Una lista de elementos duplicados.
    
    Uso del Algoritmo:

    Se define una lista numbers con algunos números.
    Se llama a detect_duplicates con esta lista.
    Se imprime la lista de números duplicados.


2. Complejidad Computacional

    O(n) + O(n) + O(n) + O(n * (n + 1)) + O(n * n) + O(n * n) + O(n) + O(1)
    = O(n) + O(n) + O(n) + O(n^2 + n) + O(n^2) + O(n^2) + O(n) + O(1)
    = O(n) + O(n) + O(n) + O(n^2) O(n) + O(n^2) + O(n^2) + O(n) + O(1)
    = O(3n^2) + O(5n) + (1)
    = O(3n^2 + 5n 1)
    = O(n^2)

    detect_duplicates:

    O(n^2) + O(1) + O(n) + O(n) + O(n) + O(1)

    Complejidad Total:

    La complejidad computacional del algoritmo completo es O(n^2).

```python
# 1. Determine el funcionamiento del algoritmo.
# 2. Calcule la complejidad computacional.

def remove_target_sum_subarrays(arr, target):
    i = 0 #O(1)
    while i < len(arr): #O(n + 1)
        j = i #O(n)
        current_sum = 0 #O(n)
        while j < len(arr): #O(n * (n + 1))
            current_sum += arr[j] #O(n * n)
            if current_sum == target: #O(n * n)
                del arr[i:j + 1] #O(n * n * n)
                i = max(0, i - 1) #O(n * n)
                break #O(n * n)
            j += 1 #O(n * n)
        else: #O(n)
            i += 1 #O(n)
    return arr #O(1)

numbers = [1, 2, 3, 7, 5, 2, 1, 2, 2, 6] #O(1)
target_sum = 5 #O(1)
result = remove_target_sum_subarrays(numbers, target_sum) # O(n^3)
print("Lista después de eliminar subarrays:", result) #O(1)
```

1. Funcionamiento del Algoritmo

    remove_target_sum_subarrays:

    Entrada: Una lista arr de números enteros y un número entero target.
    Proceso:
    Inicializa el índice i en 0.
    Utiliza un bucle while para recorrer la lista arr desde el índice i hasta el final.
    Para cada índice i, inicializa j en i y current_sum en 0.
    Utiliza un segundo bucle while para recorrer la lista desde el índice j hasta el final, sumando los elementos a current_sum.
    Si current_sum es igual a target, elimina el subarray desde i hasta j (inclusive) y ajusta i para continuar la búsqueda.
    Si no se encuentra un subarray con la suma target, incrementa i para continuar con el siguiente elemento.
    Salida: La lista arr después de eliminar todos los subarrays cuya suma es igual a target.
    
    Uso del Algoritmo:

    Se define una lista numbers con algunos números.
    Se define un target_sum de 5.
    Se llama a remove_target_sum_subarrays con numbers y target_sum.
    Se imprime la lista resultante después de eliminar los subarrays.

2. Complejidad Computacional

    remove_target_sum_subarrays:

    O(1) + O(n + 1) + O(n) + O(n) + O(n * (n + 1)) + O(n * n) + O(n * n) + O(n * n * n) + O(n * n) + O(n * n) + O(n * n) + O(n) + O(n) + O(1)
    = O(1) + O(n) + O(1) + O(n) + O(n) + O(n^2 + n) + O(n^2) + O(n^2) + O(n^3) + O(n^2) + O(n^2) + O(n^2) + O(n) + O(n) + O(1)
    = O(1) + O(n) + O(1) + O(n) + O(n) + O(n^2) + O(n) + O(n^2) + O(n^2) + O(n^3) + O(n^2) + O(n^2) + O(n^2) + O(n) + O(n) + O(1)
    = O(n^3) +  O(6n^2) + O(6n) + O(3)
    = O(n^3 + 6n^2 + 6n + 3)
    = O(n^3)

    Complejidad Total: 
    
    O(1) + O(1) + O(n^3) + O(1)
    = O(n^3) + O(3)
    = O(n^3)

    La complejidad computacional del algoritmo completo es O(n^3).


```python
# Problema: Detección de Picos en Señales de Sensores
# El objetivo es encontrar los picos (puntos máximos locales) en un conjunto de lecturas de un sensor.
# Un pico es un punto que es mayor que sus vecinos.

# 1. Calcule la complejidad computacional.
# 2. Optimice el el algoritmo utilizando recursividad.

def detect_peaks_naive(data): 
    peaks = [] #O(1)
    n = len(data) #O(1)
    for i in range(1, n - 1): #O(n)
        if data[i] > data[i - 1] and data[i] > data[i + 1]: #O(n)
            peaks.append((i, data[i])) #O(n)
    return peaks #O(1)

sensor_data = [1, 3, 7, 1, 2, 6, 3, 4, 5, 0, 8] #O(1)
peaks = detect_peaks_naive(sensor_data) #O(n)
print("Picos detectados:", peaks) #O(1)
```
